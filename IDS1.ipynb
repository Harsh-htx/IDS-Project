# Statistical Analysis

## Uttarakhand Statistical Analysis

### Scatter Plot of [SO2, NO2, PM] x [SO2, NO2, PM]

```
df = pd.DataFrame({
    "SO2": list(df3["SO2"]),
    "NO2": list(df3["NO2"]),
    "PM": list(df3["PM"])
})
df = df.dropna()
pd.plotting.scatter_matrix(df)
```

### Heat Map of Correlation Matrix

```
corrMatrix = df.corr()
sns.heatmap(corrMatrix, annot=True)
plt.show()
```

### ANOVA Testing

```
df = pd.DataFrame({
   "City": list(df3["City/Town/Village/Area"]),
    "SO2": list(df3["SO2"])
})
df = df.dropna()
```

Testing for Normality

```
sns.distplot(df['SO2'])
```

Since this follows normality we can continue with ANOVA testing

*   H0 = Mean of SO2 levels of all cities in Uttarakhand is same 
*   H1 = Mean of SO2 levels of at least 2 cities in Uttarakhand is different

```
cities = list(set(df["City"].tolist()))
city_pm = []

for i in cities:
  temp = df[df["City"]==i]
  temp = temp["SO2"].tolist()
  city_pm.append(temp)

f_stat, p_value = scipy.stats.f_oneway(*city_pm)
print("F statistic = " + str(f_stat)) 
print("P value = " + str(p_value) )
```

```
#alpha = 0.05

alpha = 0.05
print("alpha = " + str(alpha))
```

```
if p_value < alpha:
  print("H0 is rejected")
else:
  print("H0 is accepted") 
```

### Scatter Plot of [SO2, NO2, PM] x [SO2, NO2, PM]

```
df = pd.DataFrame({
    "SO2": list(df5["SO2"]),
    "NO2": list(df5["NO2"]),
    "PM": list(df5["PM"])
})
df = df.dropna()
pd.plotting.scatter_matrix(df)
```

### Heat Map of Correlation Matrix

```
corrMatrix = df.corr()
sns.heatmap(corrMatrix, annot=True)
plt.show()
```

### ANOVA Testing

```
df = pd.DataFrame({
   "City": list(df5["City/Town/Village/Area"]),
    "PM": list(df5["PM"])
})
df = df.dropna()
```

Testing for normality

```
sns.distplot(df['PM'])
```

Since this follows normality we can continue with testing

*   H0 = Mean of SO2 levels of all cities in Uttar Pradesh is same 
*   H1 = Mean of SO2 levels of at least 2 cities in Uttar Pradesh is different

```
cities = list(set(df["City"].tolist()))
city_pm = []

for i in cities:
  temp = df[df["City"]==i]
  temp = temp["PM"].tolist()
  city_pm.append(temp)

f_stat, p_value = scipy.stats.f_oneway(*city_pm)
print("F statistic = " + str(f_stat)) 
print("P value = " + str(p_value) )
```

```
#alpha = 0.05

alpha = 0.05
print("alpha = " + str(alpha))
```

### T-Test

```
df = pd.DataFrame({
   "Type": list(df5["Type of Location"]),
    "PM": list(df5["PM"])
})
df = df.dropna()
industrial = pd.DataFrame(df[df['Type']=='Industrial Area'])
residential = pd.DataFrame(df[df['Type']=='Residential, Rural and other Areas'])
```

Testing for normality

```
sns.distplot(residential['PM'])
```

```
sns.distplot(industrial['PM'])
```

Since these follow normality we can continue with t-testing

*   H0 = Mean PM levels is same for industrial and residential area
*   H1 = Mean PM level is greater for industrial than of residential area

```
industrial_sample = industrial.sample(n=50)
residential_sample = residential.sample(n=50)
```

```
industrial_sample_mean = industrial_sample['PM'].mean()
industrial_sample_std = industrial_sample['PM'].std()
industrial_sample_size = industrial_sample['PM'].count()

residential_sample_mean = residential_sample['PM'].mean()
residential_sample_std = residential_sample['PM'].std()
residential_sample_size = residential_sample['PM'].count()

t_stat_numerator = (industrial_sample_mean - residential_sample_mean)
t_stat_denominator = math.sqrt(((industrial_sample_std*industrial_sample_std)/industrial_sample_size) + ((residential_sample_std*residential_sample_std)/residential_sample_size))
t_stat = t_stat_numerator/t_stat_denominator

degree_f_numerator = (((industrial_sample_std*industrial_sample_std)/industrial_sample_size)+((residential_sample_std*residential_sample_std)/residential_sample_size))
degree_f_denominator = ((industrial_sample_std*industrial_sample_std)/(industrial_sample_size*(industrial_sample_size-1)))
degree_f_denominator = degree_f_denominator + (residential_sample_std*residential_sample_std)/(residential_sample_size*(residential_sample_size-1))
degree_f = degree_f_numerator/degree_f_denominator

print("T-stat = " + str(t_stat))
print("Degree of freedom = " + str(degree_f))
```

```
alpha = 0.05
#t_crit at df = 49 is 1.677
t_crit =1.677

print("alpha = " + str(alpha))
print("T-critical = " + str(t_crit))
```

```
if t_stat < t_crit:
  print("Failed to to reject H0")
else:
  print("H0 is rejected")
```
